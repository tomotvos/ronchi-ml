### Upload your dataset (manifest + images)
Put images under `ronchi-ml/data/images/` and your `manifest.jsonl` in `ronchi-ml/data/` with:
```json
{"id":"ronchi_4.0_-0.25_1.0","image":"data/images/example.png","meta":{"f":4.0,"offset":-0.25,"lpi":100},"labels":{"p_corr":1.0}}
```

````
#@title 3) Upload manifest.jsonl (use Files sidebar for large image folders)
from google.colab import files
up = files.upload()
open('data/manifest.jsonl','wb').write(next(iter(up.values())))
!head -n 3 data/manifest.jsonl
````

````
#@title 4) Train
!sed -n '1,200p' config.yaml
!python -m src.train --config config.yaml
````

````
#@title 5) Inference (CLI)
IMAGE_PATH='data/images/example.png'
F=4.0
OFF=-0.25
LPI=100
!python -m src.infer --image $IMAGE_PATH --f $F --offset $OFF --lpi $LPI --weights weights/best.pt
````
